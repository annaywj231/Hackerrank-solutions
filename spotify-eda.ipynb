```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
tracks = pd.read_csv('/kaggle/input/spotify-datasets/tracks.csv')
tracks.head()
```

```python
# null values
pd.isnull(tracks).sum()
```

```python
tracks.info()
```

```python
sorted_df = tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
tracks.describe().transpose()
```

```python
most_popular = tracks.query('popularity > 90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
tracks.set_index('release_date', inplace = True)
tracks.index = pd.to_datetime(tracks.index)
tracks.head()
```

```python
tracks[["artists"]].iloc[18]
```

```python
tracks["duration"] = tracks["duration_ms"].apply(lambda x: round(x/1000))
tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
tracks.duration.head()
```

```python
corr_df = tracks.drop(["key", "mode", "explicit"], axis = 1).corr(method = "pearson")
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt = ".1g", vmin = -1, vmax = 1, center = 0, cmap = "inferno", linewidths = 1, linecolor = "Black")
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
sample = tracks.sample(int(0.004 * len(tracks)))
print(len(sample))
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample, x = "energy", y = "loudness", color = "b").set(title = "Loudness VS Energy")
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample, x = "acousticness", y = "popularity", color = "r").set(title = "Popularity VS Acousticness")
```

```python
tracks["dates"] = tracks.index.get_level_values("release_date")
tracks.dates = pd.to_datetime(tracks.dates)
years = tracks.dates.dt.year
sns.displot(years, discrete = True, aspect = 2, height = 5, kind = "hist").set(title = "No. of songs per year")
```

```python
total = tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total, ax = ax, errwidth = False).set(title = "Years VS Duration")
plt.xticks(rotation = 90)
```

```python
total = tracks.duration
sns.set_style(style = "whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total, ax = ax).set(title = "Year VS Duration")
plt.xticks(rotation = 60)
```

```python
genre = pd.read_csv('/kaggle/input/ultimate-spotify-tracks-db/SpotifyFeatures.csv')
genre.head()
```

```python
plt.title("Duration in Different Genres")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(x = 'duration_ms', y = 'genre', data = genre)
plt.xlabel("Duration in ms")
plt.ylabel("Genres")
```

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize = (10,5))
well_known = genre.sort_values("popularity", ascending = False).head(10)
sns.barplot(x = 'popularity', y = 'genre', data = well_known).set(title = "Top 5 Popular Genres")
```
